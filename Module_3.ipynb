```python
import pandas as pd

# 1. Load the Data
# NOTE: Make sure Employee.csv is uploaded to Colab or update this path!
df = pd.read_csv("/content/Employee.csv")

# 2. Clean the Columns (Fixes the KeyError: 'Gender' issue)
# This removes any hidden accidental spaces at the beginning or end of column names
df.columns = df.columns.str.strip()

print("--- DATAFRAME INFORMATION ---")
df.info() # displays a complete summary of the data frame
print("\nShape:", df.shape) # returns the number of rows and columns
print("\nColumns:", df.columns) # shows all columns
print("\nIndex:", df.index) # displays row labels (index)
print("\nData Types:\n", df.dtypes) # shows the data type of each column
print("\nSize:", df.size) # returns the total number of elements
print("\nDimensions:", df.ndim) # returns the number of dimensions
print("\nMemory Usage:\n", df.memory_usage()) # displays memory used by each column
print("\nDescribe (Numeric):\n", df.describe()) # provides statistical summary of numeric columns
print("\nDescribe (All):\n", df.describe(include='all')) # shows stats for all columns
print("\nCount:\n", df.count()) # counts non-missing values in each column
print("\nMissing Values:\n", df.isnull().sum()) # counts missing values
print("\nUnique Values:\n", df.nunique()) # shows the number of unique values in each column
print("\nValue Counts (Unique Rows):\n", df.value_counts()) # counts unique rows


print("\n--- SELECTING COLUMNS & ROWS ---")
print("\nSingle Column (Gender):\n", df["Gender"])
print("\nMultiple Columns:\n", df[["Education", "City"]])

# Selecting Rows using Index (iloc uses integer positions)
print("\nFirst Row (iloc):\n", df.iloc[0])
print("\nMultiple Rows (iloc):\n", df.iloc[0:5])
print("\nLast Five Rows (iloc):\n", df.iloc[-5:])

# Selecting Rows and Columns using loc[] (uses row labels/names)
print("\nFirst Row (loc):\n", df.loc[0])
print("\nMultiple Rows (loc):\n", df.loc[0:5])


print("\n--- BOOLEAN INDEXING & FILTERING ---")
print("\nAge > 30:\n", df[df["Age"] > 30])
print("\nAge < 30:\n", df[df["Age"] < 30])
print("\nCity is Bangalore:\n", df[df["City"] == "Bangalore"])
print("\nEducation is Masters:\n", df[df["Education"] == "Masters"])

# Multiple conditions (AND - requires &)
print("\nPune AND Age > 30:\n", df[(df["City"] == "Pune") & (df["Age"] > 30)])

# Multiple conditions (OR - requires |)
print("\nPune OR Age > 30:\n", df[(df["City"] == "Pune") | (df["Age"] > 30)])

# NOT condition (requires ~)
print("\nNOT New Delhi:\n", df[~(df["City"] == "New Delhi")])

# Using isin()
print("\nCity is Bangalore or Pune:\n", df[df["City"].isin(["Bangalore", "Pune"])])

# Using between()
print("\nAge between 25 and 35:\n", df[df["Age"].between(25, 35)])

# Using str.contains() (na=False prevents errors if there are empty rows)
print("\nEducation contains 'Bachelor':\n", df[df["Education"].str.contains("Bachelor", case=False, na=False)])

# Using query() (a cleaner way to write filters)
print("\nQuery: Age > 30:\n", df.query("Age > 30"))
print("\nQuery: City == 'Pune':\n", df.query("City == 'Pune'"))
print("\nQuery: Age > 30 and Experience > 3:\n", df.query("Age > 30 and ExperienceInCurrentDomain > 3"))

# Largest and Smallest
print("\nTop 5 Oldest (nlargest):\n", df.nlargest(5, "Age"))
print("\nTop 5 Youngest (nsmallest):\n", df.nsmallest(5, "Age"))
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Aman", "Riya", "Karan", "Priya", np.nan],
    "Age": [22, np.nan, 25, 28, 24],
    "City": ["Delhi", "Mumbai", np.nan, "Lucknow", "Delhi"],
    "Salary": [50000, 60000, np.nan, 75000, 55000]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)
print(df.isnull()) #. used to identify missing value
print(df.isnull().sum()) # count missing value
print(df.notnull()) #. used to identify missing value
# fill with a constant
filled = df.fillna("Unknown")
print(filled)
# fill numeric coulmns
df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)
# fill multiple columns
df.fillna({
    "Name": "Not Available",
    "City": "Unknown",
    "Salary": 0
}, inplace=True)

print(df)

df.dropna() # removes missing values
# df.replace(old_value, new_value) # used to replace speific values (not limited to missing values)
```

# **Module-6 Data Cleaning**

**1.Duplicated()**

find duplicate values

```python
import pandas as pd

data = {
    "ID": [101, 102, 103, 104, 105],
    "Name": ["Amit", "Priya", "Rahul", "Rahul", "Neha"],
      "City": ["Delhi", "Mumbai", "Pune", "Lucknow", "Delhi"]
}

df = pd.DataFrame(data)
print("--- Original DataFrame ---")
print(df)

print("\n--- Duplicate Rows (Boolean Mask) ---")

print(df.duplicated())


print("\n--- Filtered Duplicate Data ---")
print(df[df.duplicated()])
```

**2.drop_duplicates()**

remove duplicates rows .

```python
import pandas as pd

data={
    "ID":[101,102,103,104,105],
    "Name":["Amit","Priya","Rahul","Rahul","Neha"],
    "City":["Delhi","Mumbai","Pune","Lucknow","Delhi"]
}
df=pd.DataFrame(data)
print("Original DataFrame")
print(df)

df.drop_duplicates(inplace=True)
print("\nDataFrame after removing duplicates")
print(df)
```

**3.Rename()**

Rename column names.

```python
import pandas as pd
data={
    "Emp_ID":[101,102,103,104,105],
    "Name":["Amit","Priya","Rahul","Rahul","Neha"],
    "Salary":[50000,60000,55000,70000,52000]
}
df=pd.DataFrame(data)
print("Original DataFrame")
print(df)
df=df.rename(columns={
    "Emp_ID":"ID",
    "Name":"Employee Name",
    "Salary":"Employee Salary"
})
print("\nDataFrame after renaming columns")
print(df)
```

4.astype()

Change the data type of a column.

```python
import pandas as pd
data={
    "Emp_ID":[101,102,103,104,105],
    "Salary":[50000,60000,55000,70000,52000]
}
df=pd.DataFrame(data)
print("Original DataFrame")
print(df.dtypes)
df["Emp_ID"]=df["Emp_ID"].astype(str)
print("\nDataFrame after changing data type")
print(df.dtypes)
print(df)
```

**5.replace()**

Replace value in a DataFrame.

```python
import pandas as pd

data = {
    "ID" : ["101", "102", "103"],
    "Salary" : ["45000", "60000", "55000"]
}

df = pd.DataFrame(data)

print("Before Conversion:")
print(df.dtypes)

df["ID"] = df["ID"].astype(int)
df["Salary"] = df["Salary"].astype(float)

print("\nAfter Conversion:")
print(df.dtypes)

print(df)
```

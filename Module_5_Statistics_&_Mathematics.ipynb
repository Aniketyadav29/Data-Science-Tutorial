# **1.Mean**

average value of all values Formula Mean =sum of X/n

a.sum of X= sum of all observation .
b.n=number of observation .

```python
import pandas as pd
import numpy as np


data = {
    "Employee": [
        "A", "B", "C", "D", "E",
        "F", "G", "H", "I", "J"
    ],
    "Age": [
        22, 25, 28, 30, 35,
        40, 32, 29, 45, 38
    ],
    "Salary": [
        25000, 32000, 40000, 42000, 50000,
        62000, 47000, 43000, 75000, 68000
    ],
    "Experience": [
        1, 2, 4, 5, 8,
        12, 6, 5, 18, 15
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Employee Dataset")
print(df)

# =====================================
# Mean of All Numerical Columns
# =====================================

print("\nMean of Numerical Columns")
print(df.mean(numeric_only=True))

# =====================================
# Mean of Individual Columns
# =====================================

print("\nAverage Age:", df["Age"].mean())
print("Average Salary:", df["Salary"].mean())
print("Average Experience:", df["Experience"].mean())
```

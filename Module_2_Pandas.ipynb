**Main Data Structure in Pandas**

**1.Series (1-Dimensional)**


A Series is a one-dimensional labeled array capable of holding data of any type (integers, strings, floats, Python objects, etc.).

The Analogy: Think of a Series as a single column in a spreadsheet.

The Structure: It consists of two main components: an array of actual data, and an associated array of data labels, called its index.

**2.DataFrame (2-Dimensional)**


A DataFrame is a two-dimensional, size-mutable tabular data structure with labeled axes (rows and columns). It is the most commonly used pandas object.

The Analogy: Think of a DataFrame as an entire Excel spreadsheet or a SQL database table.

The Structure: It consists of data organized into rows and columns. Under the hood, a DataFrame is essentially a collection of Series objects that all share the same index.

```python
# Creating a series with list.

import pandas as pd
data = [10, 20, 30, 40]
my_series = pd.Series(data)

print(my_series)
```

```python
# creating a series of string .
data = ['apple', 'banana', 'cherry', 'date']
my_series = pd.Series(data)

print(my_series)

```

```python
# Creating a series with custom index.
data = [10, 20, 30, 40]
index = ['a', 'b', 'c', 'd']
my_series = pd.Series(data, index=index)

print(my_series)
```

```python
# Create a series from dict.

my_dict = {'Apples': 5, 'Oranges': 8, 'Bananas': 12}
my_series = pd.Series(my_dict)

print(my_series)
```

```python
import pandas as pd

# Creating a Series with custom string labels
my_series = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

# Access by Label (like a dictionary)
print(my_series['b'])
# Output: 20

# Access by Position (like a list)
print(my_series.iloc[1])
# Output: 20
```

```python
import pandas as pd


my_series = pd.Series([40, 20, 34, 40])
print(my_series)
print(type(my_series))
print(my_series.dtype)
print(my_series.shape)
print(my_series.size)
print(my_series.ndim)

```

# **Building DataFrame**

DataFrame is a two-dimensional labeled table consisting of rows coloums a dataFrame is used to store ,organize , manipulate and analyze tabular form data .

```python
import pandas as pd
df=pd.DataFrame()
print(df)
```

```python
import pandas as pd

Student = {
    "Age": [1, 2, 3, 4],
    "Name": ["Aman", "Ravi", "Priya", "Neha"],
    "city": ["Delhi", "Lucknow", "Noida", "Jaipur"]
}

df = pd.DataFrame(Student)
print(df)
```

```python
import pandas as pd
student={
    "Name":["Aman","Ravi","Priya"],
    "Marks":[85,90,78]
}
df=pd.DataFrame(student,index=["S1","S2","S3"])
print(df)
```

create DataFrame from a list of lists

```python
import pandas as pd
data=[
    ["Aman",20,"Delhi"],
    ["Ravi",21,"Lucknow"],
    ["Priya",22,"Noida"]
]
df=pd.DataFrame(data,columns=["Name","Age","City"])
print(df)
```

Create DataFrame from a List of Dictionaries .

```python
import pandas as pd
data=[
    {"Name":"Aman","Age":20,"City":"Delhi"},
    {"Name":"Ravi","Age":21,"City":"Lucknow"},
    {"Name":"Priya","Age":22,"City":"Noida"}
]
df=pd.DataFrame(data)
print(df)
```

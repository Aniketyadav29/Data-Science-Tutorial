**Main Data Structure in Pandas**

**1.Series (1-Dimensional)**


A Series is a one-dimensional labeled array capable of holding data of any type (integers, strings, floats, Python objects, etc.).

The Analogy: Think of a Series as a single column in a spreadsheet.

The Structure: It consists of two main components: an array of actual data, and an associated array of data labels, called its index.

**2.DataFrame (2-Dimensional)**


A DataFrame is a two-dimensional, size-mutable tabular data structure with labeled axes (rows and columns). It is the most commonly used pandas object.

The Analogy: Think of a DataFrame as an entire Excel spreadsheet or a SQL database table.

The Structure: It consists of data organized into rows and columns. Under the hood, a DataFrame is essentially a collection of Series objects that all share the same index.

```python
# Creating a series with list.

import pandas as pd
data = [10, 20, 30, 40]
my_series = pd.Series(data)

print(my_series)
```

```python
# creating a series of string .
data = ['apple', 'banana', 'cherry', 'date']
my_series = pd.Series(data)

print(my_series)

```

```python
# Creating a series with custom index.
data = [10, 20, 30, 40]
index = ['a', 'b', 'c', 'd']
my_series = pd.Series(data, index=index)

print(my_series)
```

```python
# Create a series from dict.

my_dict = {'Apples': 5, 'Oranges': 8, 'Bananas': 12}
my_series = pd.Series(my_dict)

print(my_series)
```

```python
# First Graph using Matplotlib

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# Create Figure
plt.figure(figsize=(8, 5))

# Plot Line Graph
plt.plot(
    x,
    y,
    marker='o',
    linestyle='-',
    linewidth=2,
    color='blue',
    label='Sales'
)

# Title and Labels
plt.title("First Line Graph", fontsize=16)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values", fontsize=12)


# Save the graph
plt.savefig("first_graph.png", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.jpg", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.pdf", bbox_inches="tight")

# Display the graph
plt.show()

# Close the figure
plt.close()
```

# **Histogram**

A Histogram is a grapphical representation used to displey the frequency distribution of continuous numerical data .

Dataset house prices dataset Topics .

Hist()

bins()

density

cumulative Histogram

Transparency

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset

df = pd.read_csv("/content/house_prices.csv", on_bad_lines='skip', engine='python')

print("DataFrame Columns:", df.columns)

# =====================================================
# 1. Basic Histogram
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(df["Price (in rupees)"]) # A histogram groups numerical data into intervals (bins) and displays the frequency.

plt.title("Basic Histogram")

plt.show()

# =====================================================
# 2. bins
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(                         # bins decides how many intervals (groups) the data is divided into.
    df["Price (in rupees)"],
    bins=20
)

plt.title("Histogram using Bins")

plt.show()

# =====================================================
# 3. density
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(
    df["Price (in rupees)"], # Instead of showing frequency, it shows the Probability Density.
    bins=20,
    density=True
)

plt.title("Density Histogram")

plt.show()

# =====================================================
# 4. cumulative histogram
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(
    df["Price (in rupees)"],  # Shows the cumulative frequency.
    bins=20,
    cumulative=True
)

plt.title("Cumulative Histogram")

plt.show()

# =====================================================
# 5. transparency (alpha)
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(
    df["Price (in rupees)"], # Transparency is controlled using the alpha parameter.
    bins=20,
    alpha=0.5
)

plt.title("Histogram using Transparency")

plt.show()
```

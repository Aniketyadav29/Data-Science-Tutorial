**Operator**

1.Arithmetic Operator

Arithmetic operator are used to perform basic mathematical operations like , addition , substraction , multiplication and division .

```python
a=5
b=2
print("Addition",a+b)
print("Substraction",a-b)
print("Multiplication",a*b)
print("Division",a/b)
print("Modulus",a%b)
print("Exponantial",a**b)
print(a//b)
```

2.Comparision Operator

comparision operator or Relational Operator campares values. It either returns True or False accrding to the condition.

```python
a=13
b=33
print(a>b)
print(a<b)
print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
```

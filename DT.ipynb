Arithmetic operator are used to perform basic mathematical operations like , addition , substraction , multiplication and division .

4.Bitwise Operator

A bitwise operator is an operator that performs operations directly on the binary (bit) representation of integers. It manipulates individual bits (0s and 1s) of a number.


```python
a=4
b=10
print(~a)
print(a<<b)
print(a&b)
print(a^b)
print(a|b)
print(a>>b)
```

6.Identity OPerator

An identity operator is used to check whether two variables refer to the same object in memory

| Operator | Meaning                                                     |
| -------- | ----------------------------------------------------------- |
| `is`     | Returns `True` if both variables refer to the same object   |
| `is not` | Returns `True` if both variables refer to different objects |


```python
a=10
b=20
c=a
print(a is not b)
print(a is c )
```

7.Membership operator

A membership operator is used to check whether a value exists in a sequence such as a string, list, tuple, or set.

| Operator | Meaning                                                    |
| -------- | ---------------------------------------------------------- |
| `in`     | Returns `True` if the value is present in the sequence     |
| `not in` | Returns `True` if the value is not present in the sequence |


```python
x=24
y=20
my_list=[10,20,30,40,50,60]
if (x not in my_list):
  print("x is NOT present in the given list")
else:
  print("x is present in the given list")

if (y in my_list):
  print("y is present in the given list")
else:
  print("y is NOT present in the given list")
```

8.Ternary Operator

A ternary operator is a shorthand way of writing an if-else statement in a single line. It evaluates a condition and returns one of two values based on whether the condition is True or False.



```python
age = 18

result = "Adult" if age >= 18 else "Minor"
print(result)  # Adult

a,b=10,20
min = a if a<b else b
print(min)
```

#**Conditional Statements**

1.If Statement

An if statement is a conditional statement that executes a block of code only when a specified condition is True.

```python
age=19
if age>18:
  print("Eligible for Voting .")
```

2.If else Statement

If else statement is used to execute one block of code when the condition is true and another block the condition is false.

```python
a=10
if age<=12:
  print("Travel for free.")
else:
  print("Pay for ticket.")
```

3.if-elif-else Statement

elif condition is used to check multiple condition in a program . it executes a block of code when its condition evaluates to True after previous conditions evaluates to False .

```python
age=25
if age<=12:
  print("Child")
elif age<=26:
  print("Teenager")
elif age<=30:
  print("Young Adult")
elif age<=60:
  print("Adult")
else:
  print("Senior Citizen")
```

4.Nested If Else Condition

Nester If Else Condition is an is-else statement placed inside another if or else block. it is used to check condition within another condition .

```python
age=70
is_member=True
if age >=60:
  if is_member:
    print("30% Senior discount!")
  else:
    print("20% senior discount.")
else:
  print("Not eligible for a senior discount.")
```

#**Loops**

1.For Loop

A for loop iterates over a sequence (list ,tuple ,set etc.)

```python
for i in range(1,6):
  print(i)
```

Example:-

```python
fruits=["Apple","Banana","Kela", "Mango","Litchi"]
for fruit in fruits:
  print(fruit)
```

2.While Loop

A while loop runs as a condition is True .

```python
count=1
while count<=5:
  print(count)
  count +=1
```

**3.Loop Control Statement**

A.Break

stops the loops immedietely .

```python
for i in range (1,10):
  if i==5:
    break
  print(i)
```

B.Continue

Skips the current iteration

```python
for i in range (1,6):
  if i==3:
    continue
  print(i)
```

C.Pass

placehonder statements

```python
for i in range (5):
  pass

print("Loop completed ")
```

4.Nested Loops

A loop inside another loop.

```python
for i in range(1,4):
  for j in range(1,4):
    print(i,j)
```
